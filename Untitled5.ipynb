```python
dataset.drop(["StandardHours"],axis=1, inplace = True)
```

```python
dataset.drop(["Over18","EmployeeNumber"],axis=1, inplace = True)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
encoder=LabelEncoder()
```

```python
for col in dataset.columns:
    if dataset[col].dtypes=='object':
        dataset[col]=encoder.fit_transform(dataset[col])
```

```python
corr = dataset.corr()
columns = np.full((corr.shape[0],), True, dtype=bool)
for i in range(corr.shape[0]):
    for j in range(i+1, corr.shape[0]):
        if corr.iloc[i,j] >= 0.9:
            if columns[j]:
                columns[j] = False
selected_columns = dataset.columns[columns]
dataset = dataset[selected_columns]
```

```python
for col in dataset.columns: 
    print(col)
```

```python
dataset.shape
```

```python
result = pd.DataFrame()
result['Attrition'] = dataset.iloc[:,1]
```

```python
result.head()
```

```python
dataset.drop(["Attrition"],axis=1, inplace = True)
```

```python
dataset.head()
```

```python
import sklearn
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(dataset.values, result.values, test_size = 0.2)
```

```python
print("X_train shape:", x_train.shape)
print("y_train shape:", y_train.shape)

print("X_test shape:", x_test.shape)
print("y_test shape:", y_test.shape)
```

```python
from sklearn import metrics
```

```python
from sklearn.linear_model import LogisticRegression  # for Logistic Regression algorithm

logr = LogisticRegression()
logr.fit(x_train,y_train)
```

```python
y_pred = logr.predict(x_test)
acc_log = metrics.accuracy_score(y_pred,y_test)
print('The accuracy of the Logistic Regression is', acc_log)
```

```python
from sklearn import svm  #for Support Vector Machine (SVM) Algorithm

sv = svm.SVC() #select the algorithm
sv.fit(x_train,y_train)
```

```python
y_pred = sv.predict(x_test) #now we pass the testing data to the trained algorithm
acc_svm = metrics.accuracy_score(y_pred,y_test)
print('The accuracy of the SVM is:', acc_svm)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

```python
model = RandomForestClassifier()
model.fit(x_train,y_train)
#predict the results for test
test_pred = model.predict(x_test)
#test the accuracy
accuracy_score(y_test, test_pred)
```

```python
feat_importances = pd.Series(model.feature_importances_, index=dataset.columns)
feat_importances = feat_importances.nlargest(20)
feat_importances.plot(kind='barh')
```
